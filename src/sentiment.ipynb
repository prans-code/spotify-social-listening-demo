```python
import nltk
nltk.download('vader_lexicon')
```

```python
import pandas as pd
from nltk.sentiment.vader import SentimentIntensityAnalyzer

INPUT_PATH = "../data/processed_reddit_spotify.csv"
OUTPUT_PATH = "../data/sentiment_reddit_spotify.csv"

def label_sentiment(score):
    if score >= 0.05:
        return "positive"
    elif score <= -0.05:
        return "negative"
    else:
        return "neutral"

if __name__ == "__main__":
    df = pd.read_csv(INPUT_PATH)

    analyzer = SentimentIntensityAnalyzer()

    df["sentiment_score"] = df["clean_text"].apply(
        lambda x: analyzer.polarity_scores(x)["compound"]
    )

    df["sentiment_label"] = df["sentiment_score"].apply(label_sentiment)

    df["sentiment_confidence"] = df["sentiment_score"].abs()
    
    df.to_csv(OUTPUT_PATH, index=False)
    print(f"Saved sentiment data to {OUTPUT_PATH}")
```
