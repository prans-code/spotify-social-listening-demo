```python
import pandas as pd

INPUT_PATH = "../data/sentiment_reddit_spotify.csv"
SUMMARY_OUTPUT = "../data/sentiment_summary.csv"
TIME_OUTPUT = "../data/sentiment_time_series.csv"

if __name__ == "__main__":
    df = pd.read_csv(INPUT_PATH)

    # Ensure datetime
    df["created_utc"] = pd.to_datetime(df["created_utc"], errors="coerce")
    df = df.dropna(subset=["created_utc"])
    df["date"] = df["created_utc"].dt.date

    # ---- Sentiment summary (FIXED) ----
    summary = (
        df
        .groupby("sentiment_label")
        .size()
        .reset_index(name="count")
        .rename(columns={"sentiment_label": "sentiment"})
    )

    summary["percentage"] = summary["count"] / summary["count"].sum() * 100

    # ---- Sentiment over time ----
    time_series = (
        df
        .groupby(["date", "sentiment_label"])
        .size()
        .reset_index(name="count")
    )

    summary.to_csv(SUMMARY_OUTPUT, index=False)
    time_series.to_csv(TIME_OUTPUT, index=False)

    print("Aggregation completed successfully.")
```
